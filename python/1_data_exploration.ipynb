# Data Quality Assessment 

## Dataset Overview

The Divvy Trips 2019 Q1 dataset contains 365,069 records and 12 columns.

Initial exploration was conducted using:
- df.shape
- df.info()
- df.isnull().sum()

The objective was to identify missing values, data type inconsistencies, and other potential data quality issues before cleaning.

## Findings

### Missing Values

The following columns contain missing values:

| Column | Missing Values |
|----------|----------|
| gender | 19,711 |
| birthyear | 18,023 |

All other columns contain no missing values.

### Data Type Issues

The following columns were identified as having potentially incorrect data types:

| Column | Current Type | Expected Type |
|----------|----------|----------|
| start_time | object (string) | datetime |
| end_time | object (string) | datetime |
| tripduration | object (string) | numeric |

### Trip Duration Investigation

Further investigation revealed that the tripduration column contains values formatted with comma separators (e.g., 1,783.0).

Approximately 72,042 records contain comma-separated values, which likely caused Pandas to interpret the column as a string rather than a numeric datatype.

### Schema Consistency Observation

The 2019 and 2020 datasets use different schemas.

The 2019 dataset contains demographic fields such as:
- gender
- birthyear

These fields are not present in the 2020 dataset and may require special consideration during data integration.

```python
# Data Cleaning Function for 2019 Data set

def clean_2019_data(df):
    
    # Create working copy
    df = df.copy()
    
    # Convert date columns
    df['start_time'] = pd.to_datetime(df['start_time'])
    df['end_time'] = pd.to_datetime(df['end_time'])
    
    # Clean tripduration

    ## Remove commas
    df['tripduration'] = (
        df['tripduration']
        .str.replace(',', '', regex=False)
    )
    
    ## convert tripduration column
    df['tripduration'] = pd.to_numeric(df['tripduration'])
    
    return df
```

```python
# Clean q1 data set
q1_2019_clean = clean_2019_data(df)

# Clean q2 data set
q2_2019_clean = clean_2019_data(df_q2_clean)

# Clean q3 data set
q3_2019_clean = clean_2019_data(df_q3_clean)

# Clean q4 data set
q4_2019_clean = clean_2019_data(df_q4_clean)
```
